## Part 1: NumPy - Numerical Python

NumPy is the foundation for scientific computing in Python. It provides:
- Fast operations on arrays and matrices
- Mathematical functions
- Random number generation

```python
# Import NumPy (standard alias: np)
import numpy as np
```

### Creating Arrays

Arrays are the fundamental data structure in NumPy.

```python
# 1D array (vector)
vector = np.array([1.2, 2.4, 3.5, 4.7, 6.1])
print("1D array:")
print(vector)
```

```python
# 2D array (matrix)
matrix = np.array([[6, 5], [11, 7], [4, 8]])
print("2D array:")
print(matrix)
print(f"\nShape: {matrix.shape}")  # (rows, columns)
```

### Generating Arrays

```python
# Sequence of numbers (like Python's range)
sequence = np.arange(5, 12)  # start (inclusive) to stop (exclusive)
print("Sequence:", sequence)
```

```python
# Arrays of zeros and ones
zeros = np.zeros(5)
ones = np.ones((3, 4))  # 3x4 matrix of ones
print("Zeros:", zeros)
print("\nOnes:\n", ones)
```

### Random Numbers

```python
# Random integers
random_ints = np.random.randint(low=50, high=101, size=6)
print("Random integers [50-100]:", random_ints)
```

```python
# Random floats [0.0, 1.0)
random_floats = np.random.random(6)
print("Random floats [0-1]:", random_floats)
```

### Mathematical Operations

NumPy uses **broadcasting** to apply operations to entire arrays efficiently.

```python
# Add a constant to all elements
data = np.array([1, 2, 3, 4, 5])
result = data + 10
print("Original:", data)
print("After +10:", result)
```

```python
# Multiply all elements
result = data * 3
print("After *3:", result)
```

```python
# Element-wise operations between arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("a + b =", a + b)
print("a * b =", a * b)
```

### Task 1: Create a Linear Dataset

Create two arrays:
1. `feature`: integers from 6 to 20 (inclusive) using `np.arange()`
2. `label`: calculated as `label = 3 * feature + 4`

**Hint:** Remember that `np.arange(start, stop)` is exclusive of the stop value.

```python
# Task 1: Write your code here
feature = np.arange(6, 21)  # Create array from 6 to 20 (inclusive)
label = 3 * feature + 4    # Calculate: 3 * feature + 4

print("Feature:", feature)
print("Label:", label)
```

### Task 2: Add Noise to the Dataset

To make your dataset more realistic, add random noise:

1. Create a `noise` array with the same shape as `label`
2. Fill it with random floating-point values between -2 and +2
3. Add the noise to your `label` array to create `noisy_label`

**Hints:**
- Use `np.random.random()` to generate values between 0 and 1
- Scale and shift these values to get the range [-2, +2]
- Formula: `random_value * range + minimum`

```python
# Task 2: Write your code here
noise = np.random.random() * 2 - 2  # Create noise array between -2 and +2
noisy_label = label * noise  # Add noise to label

print("Original label:", label)
print("Noise:", noise)
print("Noisy label:", noisy_label)
```

### Task 3: Array Statistics

NumPy provides functions to calculate statistics. Calculate the following for the `noisy_label` array:

1. Mean (average) value
2. Maximum value
3. Minimum value
4. Standard deviation

**Hints:** Use `np.mean()`, `np.max()`, `np.min()`, and `np.std()`

```python
# Task 3: Write your code here
mean_value = np.mean(noisy_label)      # Calculate mean
max_value = np.max(noisy_label)        # Calculate maximum
min_value = np.min(noisy_label)        # Calculate minimum
std_value = np.std(noisy_label)        # Calculate standard deviation

print(f"Mean: {mean_value}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")
print(f"Std: {std_value}")
```

### Task 6: Filter by Gender

Create two new DataFrames:
1. `female_passengers`: only female passengers
2. `male_passengers`: only male passengers

Then print how many passengers are in each group.

**Hint:** Use boolean indexing like `df[df['column'] == value]`

```python
# Task 6: Write your code here
female_passengers = titanic[titanic['Sex'] == 'female']    # Filter for Sex == 'female'
male_passengers = titanic[titanic['Sex'] == 'male']     # Filter for Sex == 'male'

print(f"Female passengers: {len(female_passengers)}")
print(f"Male passengers: {len(male_passengers)}")
```

### Task 7: Calculate Survival Rate by Gender

Calculate the percentage of survivors for each gender:
1. For female passengers, calculate: (number who survived / total females) × 100
2. For male passengers, calculate: (number who survived / total males) × 100

**Hints:**
- Use `.sum()` on the 'Survived' column to count survivors (1 = survived, 0 = died)
- Use `len()` to get total number of passengers

```python
# Task 7: Write your code here
female_survival_rate = (female_passengers['Survived'].sum() / len(female_passengers)) * 100   # Calculate percentage
male_survival_rate = (male_passengers['Survived'].sum() / len(male_passengers)) * 100      # Calculate percentage

print(f"Female survival rate: {female_survival_rate:.1f}%")
print(f"Male survival rate: {male_survival_rate:.1f}%")
```

### Task 8: Create a New Column

Create a new column called 'Family_Size' that combines:
- 'Siblings/Spouses Aboard'

- Plus 1 (for the passenger themselves)

Then display the first 10 rows showing Name, Family_Size, and Survived.

### Task 9: Find the Most Expensive Ticket

Find information about the passenger(s) who paid the highest fare:
1. Find the maximum fare value
2. Filter the DataFrame to show only rows where Fare equals this maximum
3. Display the Name, Pclass, and Fare columns

**Hints:**
- Use `.max()` to find the maximum value
- Use boolean indexing to filter

```python
# Task 9: Write your code here
max_fare = titanic['Fare'].max()                      # Find maximum fare
most_expensive = titanic[titanic['Fare'] == max_fare]                # Filter for rows with max fare

print(f"Maximum fare: ${max_fare:.2f}")
print("\nPassenger(s) with most expensive ticket:")
most_expensive[['Name', 'Pclass', 'Fare']]
```

### Task 10: Group by Class and Calculate Average Age

Calculate the average age for each passenger class (Pclass):
1. Group the data by 'Pclass'
2. Calculate the mean of 'Age' for each group

**Hint:** Use `.groupby('column')['other_column'].mean()`

```python
# Task 10: Write your code here
average_age_by_class = titanic.groupby(titanic['Pclass'])['Age'].mean()    # Group by Pclass and calculate mean Age

print("Average age by passenger class:")
print(average_age_by_class)
```

### Task 11: Multiple Conditions Filter

Create a DataFrame of passengers who meet ALL these conditions:
- Female
- First class (Pclass == 1)
- Age less than 30

Count how many passengers meet these criteria.

**Hint:** Combine conditions with `&` (and): `df[(condition1) & (condition2) & (condition3)]`

```python
# Task 11: Write your code here
filtered_passengers = titanic[(titanic['Sex'] == 'female') & (titanic['Pclass'] == 1) & (titanic['Age'] < 30)]    # Apply all three conditions

print(f"Number of young first-class female passengers: {len(filtered_passengers)}")
print("\nFirst few:")
filtered_passengers[['Name', 'Age', 'Pclass', 'Sex']].head()
```

### Task 12: Missing Data Analysis

Check for missing data in the Titanic dataset:
1. Count how many missing values are in each column
2. Calculate the percentage of missing values for the 'Age' column

**Hints:**
- Use `.isnull()` to create a boolean mask of missing values
- Use `.sum()` to count True values
- Apply to each column with `.isnull().sum()`

```python
# Task 12: Write your code here
missing_counts = titanic.isnull().sum()             # Count missing values per column
age_missing_percent = titanic['Age'].isnull().sum() / len(titanic) * 100        # Calculate percentage for Age column

print("Missing values per column:")
print(missing_counts)
print(f"\nPercentage of missing Age values: {age_missing_percent:.1f}%")
```
